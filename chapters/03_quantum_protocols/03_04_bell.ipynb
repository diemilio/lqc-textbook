# Bell Inequalities
<hr></hr>

In 1935, Albert Einstein, Boris Podolsky, and Nathan Rosen published a paper that would set the stage for one of the most debated ideas in science history: [quantum entanglement](https://en.wikipedia.org/wiki/Quantum_entanglement). In their work entitled *"Can Quantum-Mechanical Description of Physical Reality Be Considered Complete?"* (now referred as the EPR paper after their initials), the authors point out that quantum theory presents an inconsistency. By the use of a thought experiment, they show that quantum mechanics alone cannot explain the results predicted in scenarios involving entangled particles, and is therefore an **incomplete** theory [[Einstein35](https://journals.aps.org/pr/pdf/10.1103/PhysRev.47.777)]. In other words, EPR argues that for quantum mechanics to be correct, it has to be complemented with what are now often referred as [local hidden-variables](https://en.wikipedia.org/wiki/Local_hidden-variable_theory).

This issue remained unresolved for almost 30 years, when in 1964 [John Stewart Bell](https://en.wikipedia.org/wiki/John_Stewart_Bell) published his revolutionary paper *"On the Einstein Podolsky Rosen Paradox"*. In it, Bell demonstrated that quantum mechanics was indeed complete, and therefore did not require additional local hidden-variables [[Bell64](https://journals.aps.org/ppf/pdf/10.1103/PhysicsPhysiqueFizika.1.195)].

In this chapter, we will explain the importance and implications of Bell's theorem by the use the Clauser–Horne–Shimony–Holt [(CHSH) inequality](https://en.wikipedia.org/wiki/CHSH_inequality#CHSH_game), which is simply a more qubit-friendly version of Bell's original formulation. In section 1, we will frame the EPR thought experiment and Bell's argument in the form of a game (known as the [CHSH game](https://en.wikipedia.org/wiki/CHSH_inequality#CHSH_game)) that can be casted as a quantum circuit, and demonstrate that quantum mechanics does indeed predict the results seen in experimentation.

In section 2 we will present a second thought experiment known as the [GHZ-Mermin](https://en.wikipedia.org/wiki/Quantum_pseudo-telepathy#Greenberger%E2%80%93Horne%E2%80%93Zeilinger_game) game, which reinforces the validity of Bell's theorem by showing a wider gap between what quantum mechanics predicts to what can be accomplished with a classical theory.

##### Something Worth Mentioning

Many (probably most) explanations of Bell's theorem claim that what Bell really proved was that quantum mechanics is a [non-local theory](https://en.wikipedia.org/wiki/Quantum_nonlocality). What this implies is that, in a system composed of, for example, two entangled particles, measuring one particle instantaneously determines the outcome of the other, no matter how far apart they are. Even though this is a helpful perspective when working with quantum states, this isn't really what Bell demonstrated. Frameworks such as Deutsch and Hayden's [theory of descriptors](https://arxiv.org/pdf/quant-ph/9906007) can perfectly explain the effects of quantum mechanics as a local theory without violating what Bell's proof. 

Regardless, in this chapter we will take the view of standard quantum mechanics, where non-locality is unavoidable. It is just worth pointing out that there are other valid formulations where locality is preserved and still adhere to Bell's theorem (see for example: [[Bédard25](https://arxiv.org/pdf/2406.12184v2)].
