# Bell Inequalities
<hr></hr>

In 1935, Albert Einstein, Boris Podolsky, and Nathan Rosen published a paper that would set the stage for one of the most debated ideas in science history: [quantum entanglement](https://en.wikipedia.org/wiki/Quantum_entanglement). In their work entitled *"Can Quantum-Mechanical Description of Physical Reality Be Considered Complete?"* (now referred as the EPR paper after their initials), the authors point out that the theory of quantum mechanics presents a serious inconsistency. By the use of a thought experiment, they show that quantum mechanics alone cannot explain the results predicted in scenarios involving entangled particles, and is therefore an **incomplete** theory [[Einstein35](https://journals.aps.org/pr/pdf/10.1103/PhysRev.47.777)]. In other words, EPR argues that for quantum mechanics to be correct, it has to be complemented with what are now often referred as [local hidden-variables](https://en.wikipedia.org/wiki/Local_hidden-variable_theory).

This issue remained unresolved for almost 30 years, when in 1964 [John Stewart Bell](https://en.wikipedia.org/wiki/John_Stewart_Bell) published his revolutionary paper *"On the Einstein Podolsky Rosen Paradox"*. In it, Bell demonstrated that quantum mechanics was indeed complete, and therefore did not require additional local hidden-variables [[Bell64](https://journals.aps.org/ppf/pdf/10.1103/PhysicsPhysiqueFizika.1.195)].

In this chapter, we will explain the importance and implications of Bell's theorem by the use the Clauser–Horne–Shimony–Holt [(CHSH) inequality](https://en.wikipedia.org/wiki/CHSH_inequality#CHSH_game), which is simply a more qubit-friendly version of Bell's original formulation. In section 1, we will frame the EPR thought experiment and Bell's argument in the form of a game (known as the [CHSH game](https://en.wikipedia.org/wiki/CHSH_inequality#CHSH_game)) that can be casted as a quantum circuit, and demonstrate that quantum mechanics does indeed predict the results seen in experimentation.

In section 2 we will present a second thought experiment known as the [GHZ-Mermin](https://en.wikipedia.org/wiki/Quantum_pseudo-telepathy#Greenberger%E2%80%93Horne%E2%80%93Zeilinger_game) game, which reinforces the validity of Bell's theorem by showing a wider gap between what quantum mechanics predicts and what can be explained with a purely classical theory.

##### Something Worth Noting

Many (probably most) explanations of Bell's theorem claim that what Bell proved was that quantum mechanics is a [non-local theory](https://en.wikipedia.org/wiki/Quantum_nonlocality). What this implies is that, in systems composed of, for example, two entangled particles, measuring one particle instantaneously determines the outcome of the other, no matter how far apart they are. Even though this is a helpful perspective when working with quantum states, for many—including myself—this view remains controversial. Frameworks such as Deutsch and Hayden's [theory of descriptors](https://arxiv.org/pdf/quant-ph/9906007) can perfectly explain the effects of quantum mechanics as a local theory without violating what Bell's proof.

Regardless, in this chapter we will take the view of standard quantum mechanics, where non-locality is unavoidable. It is just important to point out that there are other valid formulations where locality is preserved and still adhere to Bell's theorem (see for example: [[Bédard25](https://arxiv.org/pdf/2406.12184v2)]).

## 1. The CHSH Game

Five years after Bell's seminal paper on the [EPR paradox](https://en.wikipedia.org/wiki/Einstein%E2%80%93Podolsky%E2%80%93Rosen_paradox), John Clauser, Michael Horne, Abner Shimony, and Richard Holt (CHSH), presented a straightforward version of Bell's argument that could be experimentally tested in a laboratory using photons [[Clauser69](https://journals.aps.org/prl/pdf/10.1103/PhysRevLett.23.880)]. It's simplicity then allowed researchers in the quantum information space to generalize the CHSH formulation as a game that could be adapted to the general context of qubits evolving through quantum circuits. This facilitates the demonstration of how quantum mechanics predicts the results of certain experiments involving entangled systems that cannot be explained using any type of classical theory.

The key outcome from this thought experiment is that, quantum mechanics correctly predicts that the probability of winning the CHSH gate is of $~85\%.$ On the other hand, any classical strategy—which could be use as a theory of local hidden variables to complement quantum mechanics as EPR proposed—predicts that the maximum probability of success is only $75 \%,$ which contradicts experimental data.

### 1.1 Game Setup

The rules of the CHSH game are simple. There are two players—Alice and Bob—who need to work together to win the game. It is also required for the players to be separated without the ability to communicate with each other. Therefore, Alice and Bob must coordinate and agree on a common strategy **before** the game starts.

<img src="images/03_04_01_chsh_game.png" align = "center" width="575"/>

A referee—Eve—sends bits $a$ and $b$ to Alice and Bob, respectively $($ with $a, b \in \{0,1\}).$ Eve picks which combination of bits $(a,b)$ to send at random with equal probability.

In return, each players must respond to Eve with their own bit of information; here captured by $x$ for Alice and $y$ for Bob.

To win the game, the values of $x$ and $y$ that Alice and Bob send back to Eve must be **different** from each other whenever both of the bits they received are equal to $1.$ In other words, only when $a = b = 1,$ $x \neq y;$ in all other cases, $x = y.$

We can summarize the winning conditions using the following table:

$$
\begin{array}{|c|c|c|}
\hline
a & b & \text{Winning Condition} \\
\hline
0 & 0 & x = y \\
\hline
0 & 1 & x = y \\
\hline
1 & 0 & x = y \\
\hline
1 & 1 & x \neq y \\
\hline
\end{array}
$$

Said differently, the players win if the following condition holds:

$$ a \cdot b = x \oplus y .$$

Eve then knows that Alice and Bob won when the product of the bits she sent (left-hand side of the equation) is equal to the sum—modulo 2—of the bits she received (right-hand side of the equation).

### 1.2 Classical Strategy

To maximize their probability of winning, Alice and Bob must come up with a strategy before they're separated. Since they are working with only one bit each, their options are limited. Here are the possible schemes they can use:

####  Send Eve a constant bit value.

In this strategy, Alice and Bob agree that they will always send back to Eve a predetermined bit value. For example they can coordinate so that both of them reply with a $0,$ regardless of what they receive; i.e., $ (x = 0, y = 0).$

We can put together a table for the four possible values of bits Eve can send, and see when the winning condition $ a \cdot b = x \oplus y $ is met:

$$
\begin{array}{|c|c|c|c|c|c|c|}
\hline
a & b & x & y & a \cdot b & x \oplus y & (a \cdot b = x \oplus y) \\
\hline
0 & 0 & 0 & 0 & 0 & 0 & ✔ \\
\hline
0 & 1 & 0 & 0 & 0 & 0 & ✔ \\
\hline
1 & 0 & 0 & 0 & 0 & 0 & ✔ \\
\hline
1 & 1 & 0 & 0 & 1 & 0 & ✗ \\
\hline
\end{array}
$$

With this game plan, Alice and Bob have a probability of success of $3$ in $4$, or $75\%.$

Similarly, we can calculate the probability of winning for the other three possible combinations of constant values. For the case $(x = 1, y = 1),$ we have $x \oplus y = 0,$ which gives us the same results as above. And for the cases where $(x = 0, y = 1)$ and $(x = 1, y = 0)$ we have $x \oplus y = 1,$ which is only equal to $a \cdot b = 1$ in 1 out 4 possible options. As a second example, here's the table for when Alice and Bob decide to send back $(x = 0, y = 1),$ which confirms a $25 \%$ probability of winning:

$$
\begin{array}{|c|c|c|c|c|c|c|}
\hline
a & b & x & y & a \cdot b & x \oplus y & (a \cdot b = x \oplus y) \\
\hline
0 & 0 & 0 & 1 & 0 & 1 & ✗ \\
\hline
0 & 1 & 0 & 1 & 0 & 1 & ✗ \\
\hline
1 & 0 & 0 & 1 & 0 & 1 & ✗ \\
\hline
1 & 1 & 0 & 1 & 1 & 1 & ✔ \\
\hline
\end{array}
$$

In summary, we have the following probabilities of winning for each possible combination of constant values:

| Selected $x$ | Selected $y$ | Winning Probability |
| :-: | :-: | :-: |
| $0$ | $0$ | $75\%$ |
| $0$ | $1$ | $25\%$ |
| $1$ | $0$ | $25\%$ |
| $1$ | $1$ | $75\%$ |

####  Send Eve the same (or opposite) bit value received.

This second strategy consists of sending back to Eve a value that depends on the bit received. For example, Alice and Bob could agree to simply send back the same value the get $(x = a, y = b),$ which results in the following table:

$$
\begin{array}{|c|c|c|c|c|c|c|}
\hline
a & b & x & y & a \cdot b & x \oplus y & (a \cdot b = x \oplus y) \\
\hline
0 & 0 & 0 & 0 & 0 & 0 & ✔ \\
\hline
0 & 1 & 0 & 1 & 0 & 1 & ✗ \\
\hline
1 & 0 & 1 & 0 & 0 & 1 & ✗ \\
\hline
1 & 1 & 1 & 1 & 1 & 0 & ✗ \\
\hline
\end{array}
$$

This is clearly not a great strategy since only gives the players a $25 \%$ chance of winning. 

Likewise, we can calculate the probability of winning for the other three possible combinations. In the case where both Alice and Bob send back the opposite value they receive $(x = \bar{a}, y = \bar{b}),$ the outcome is identical to the one above. Alternatively, when Alice sends to Eve the same bit value she receives, but Bob sends her the opposite $(x = a, y = \bar{b}),$ we get a probability of success of $75 \%:$ 

$$
\begin{array}{|c|c|c|c|c|c|c|}
\hline
a & b & x & y & a \cdot b & x \oplus y & (a \cdot b = x \oplus y) \\
\hline
0 & 0 & 0 & 1 & 0 & 1 & ✗ \\
\hline
0 & 1 & 0 & 0 & 0 & 0 & ✔ \\
\hline
1 & 0 & 1 & 1 & 0 & 0 & ✔ \\
\hline
1 & 1 & 1 & 0 & 1 & 1 & ✔ \\
\hline
\end{array}
$$

And, of course, the case in which Alice sends to Eve the opposite to what she receives but Bob sends her the same $(x = \bar{a}, y = b),$ produces the same results.

In summary, we have the following probabilities of winning for each possible combination of returned values:

| Selected $x$ | Selected $y$ | Winning Probability |
| :-: | :-: | :-: |
| $a$ | $b$ | $25\%$ |
| $a$ | $\bar{b}$ | $75\%$ |
| $\bar{a}$ | $b$ | $75\%$ |
| $\bar{a}$ | $\bar{b}$ | $25\%$ |

####  Send Eve bit values based on a probabilistic strategy.

One last strategy would be for Alice and Bob to pick their values at random. For example, they could decide that, with probability $\mathbb{P}_a$ Alice sends back a $0$, and probability $1 - \mathbb{P}_a$ sends back a $1.$ Similarly, with probability $\mathbb{P}_b$ Bob sends back a $0$, and probability $1 - \mathbb{P}_a$ sends back a $1.$ 

The thing about this strategy is that, at the end of the day, Alice and Bob end up sending one of the four possible constant values discussed in the first strategy, each with the following probabilities of occurrence:

- With probability $\mathbb{P}_a \mathbb{P}_b,$ Alice and Bob send to Eve $(x = 0, y = 0).$
- With probability $(1-\mathbb{P}_a) \mathbb{P}_b,$ Alice and Bob send to Eve $(x = 1, y = 0).$
- With probability $\mathbb{P}_a (1-\mathbb{P}_b),$ Alice and Bob send to Eve $(x = 0, y = 1).$
- With probability $(1-\mathbb{P}_a) (1-\mathbb{P}_b),$ Alice and Bob send to Eve $(x = 1, y = 1).$

And since we already know the probability of winning the game for each of the four deterministic options, we can compute the probability of winning for some given values of $\mathbb{P}_a$ and $\mathbb{P}_b:$ 

$$ 
\begin{aligned}
\mathbb{P}_{\text{win}} &= \frac{3}{4}\mathbb{P}_a \mathbb{P}_b + \frac{1}{4}(1-\mathbb{P}_a) \mathbb{P}_b + \frac{1}{4}\mathbb{P}_a (1-\mathbb{P}_b) + \frac{3}{4} (1-\mathbb{P}_a) (1-\mathbb{P}_b)
\\
\\
\mathbb{P}_{\text{win}} &= \frac{3}{4} - \frac{1}{2}(\mathbb{P}_a + \mathbb{P}_b) + \mathbb{P}_a \mathbb{P}_b
\end{aligned}
$$

What's key here is that, the maximum value $\mathbb{P}_{\text{win}}$ can take is $3/4$ ($75\%$), which is obtained when either $\mathbb{P}_a = \mathbb{P}_b = 1$ or $\mathbb{P}_a = \mathbb{P}_b = 0.$ This simply corresponds to the deterministic strategies obtained from always sending back either $(x = 0, y = 0)$ or $(x = 1, y = 1).$

A similar argument can be made for when, instead of selecting a constant value, Alice and Bob decide to send back the same or the opposite value they receive with some given probability (e.g. with prob $\mathbb{P}_a$ Alice sends back $a$, etc.). However, the conclusion for this approach is the same as above: the maximum probability of success for **any** classical strategy is of $75\%.$

### 1.3 Quantum Strategy
